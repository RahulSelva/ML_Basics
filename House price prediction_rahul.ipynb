```python
import pandas as pd
df=pd.read_csv('E:\Machine Learning\Sundhar sir lecture\Melbourne_housing_FULL.csv')
df=df.drop(['Address','Method','Date','Bedroom2','Distance','Lattitude','Longtitude','YearBuilt','SellerG'],axis=1)
df.dropna(subset = ['Landsize'], inplace=True)
df['Car'].fillna(value=0,inplace=True)
df['Bathroom'].fillna(value=1,inplace=True)
df
```

```python
import numpy as np
df['result'] = np.where(df['Landsize'] >= df['BuildingArea'],df['Landsize'],df['BuildingArea'])
# df
```

```python
df['Landsize'] = np.where(df['Landsize']==0 ,df['BuildingArea'],df['Landsize'])
df=df.drop('BuildingArea',axis=1)
df=df.drop('result',axis=1)
df.sort_values('Landsize',ascending=True,inplace=True)
df.dropna(subset=['Landsize'],inplace=True)
df=df.iloc[100:]
#df.sort_values('Price',ascending=True,inplace=True)
df
```

```python
df.isna().sum()
```

```python
df.sort_values('Price',ascending=False,inplace=True)
df.dropna(subset=['Price'],inplace=True)
df
```

```python

df = df[['Rooms','Propertycount','Postcode','Bathroom','Car','Landsize','Type','CouncilArea','Regionname','Suburb','Price']]
df=df.reset_index(drop=True)
df
```

```python
# new data frame with split value columns
new = df["CouncilArea"].str.split(" ", n = 1, expand = True)
new1 = df["Regionname"].str.split(" ", n = 1, expand = True)
# making separate first name column from new data frame
df["Council"]= new[0]
df["Region"]= new1[0]
  
# making separate last name column from new data frame
# df["Last Name"]= new[1]

df.drop(columns =["CouncilArea",'Regionname'], inplace = True)
df = df[['Rooms','Propertycount','Postcode','Bathroom','Car','Landsize','Type','Council','Region','Suburb','Price']]

# Dropping old Name columns
# data.drop(columns =["Name"], inplace = True)
df
```

```python
# Assigning Values
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values

# getting dummies
d1=pd.get_dummies(x[:,6])
d2=pd.get_dummies(x[:,7])
d3=pd.get_dummies(x[:,8])
d4=pd.get_dummies(x[:,9])

# Adding Dummies ,assigning to df , dropping original, concatenating, renaming, droping last column of dummies
x=pd.DataFrame(x)
x.drop([3,6,7,8,9],inplace=True,axis=1)
x=pd.concat([x,d1,d2,d3,d4],axis=1)
x.rename(columns={0:'Rooms',1:'Propertycount',2:'Bathroom',5:'Landsize',4:'Car'},inplace=True)
x.drop(['u','Yarra','Western','Yarraville'],inplace=True,axis=1)
x
```
